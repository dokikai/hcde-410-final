Below is the code I used for visualizing the data from site 2:

First, I needed to import relevant libraries and read the csv file for the first type of ART medication that I stratified into a dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the NRTI data
data_nrti = pd.read_csv('nrti_data.csv')
```

Currently, we are working in the NRTI (Nucleoside reverse transcriptase inhibitors) dataset which is a classification of drug treatment for treating HIV After doing so, I had to make a design decision as to how I wanted to represent the data as it had two separate costs that indicated a gap/markup in terms of the pharmaceutical price. So, I decided a bar/cat chart would be the best that visualizes for each drug the cost in USD before and after the markup to indicate a single step where costs are increased for profit.

```python
# Gets the number of pairs of bars we need 
N = len(data_nrti)


# Specifies blue bars (before markup)
blue_bar = data_nrti["Cheapest Wholsale Acquisition Cost"]
# Specifies the orange bars (after markup)
orange_bar = data_nrti["Average Wholesale Price"]

# Arrange and align the bars 
ind = np.arange(N)

# Figure size
plt.figure(figsize=(10,5))

# Bar width
width = 0.3       

# Plot the information
plt.bar(ind, blue_bar , width, label='Wholesale Acquisition Cost')
plt.bar(ind + width, orange_bar, width, label='Average Wholesale Price')

plt.xlabel('Drug Name')
plt.ylabel('Cost per prescription (USD)')
plt.title('NRTI HIV Treatment Wholesale Markup Diagram')


# Formatting
plt.xticks(ind + width / 2, data_nrti["Drug Name"])
plt.xticks(rotation=-80)

# Making the format ideal for the legend and saving it
plt.legend(loc='best')
plt.savefig('NRTI.png', dpi=300, bbox_inches='tight')
plt.show()
```

As the graph above indicates, there is a discrepancy with regards to the wholesale acquisition cost that vendors pay along with the average wholesale price which is used as the basis for setting reimbursement rates along with direct costs if insurance doesn't cover it. This can also be done with all other forms of ART for HIV, and in order to maximize efficiency in the code, we declare a method that can generalize this process for many different datasets. This is defined below and took in two parameters, one which will be the name of the file we are reading and the second being the name of the graph to describe the data. We also add a file_name parameter to document the name of the file that gets saved.

```python
def make_plot(csv_name, title_name, file_name):
    # Inputted parameter csv_name becomes the dataset to be read
    data_nrti = pd.read_csv(csv_name)
    N = len(data_nrti)
    blue_bar = data_nrti["Cheapest Wholsale Acquisition Cost"]
    orange_bar = data_nrti["Average Wholesale Price"]
    ind = np.arange(N)
    plt.figure(figsize=(10,5))
    width = 0.3       

    # title_name becomes the title of the graph
    plt.bar(ind, blue_bar , width, label='Wholesale Acquisition Cost')
    plt.bar(ind + width, orange_bar, width, label='Average Wholesale Price')
    plt.xlabel('Drug Name')
    plt.ylabel('Cost per prescription (USD)')
    plt.title(title_name)
    plt.xticks(ind + width / 2, data_nrti["Drug Name"])
    plt.xticks(rotation=-80)
    plt.legend(loc='best')
    plt.savefig(file_name, dpi=300, bbox_inches='tight')
    plt.show()
```

After defining the function, we then pass in different parameters in order to visualize each dataset we collected. That is, we want to visualize the treatment methods (in order) of Non-nucleoside reverse transcriptase inhibitors (NNRTI), combinations of Nucleoside reverse transcriptase inhibitors, Protease inhibitor treatment, and a combination of other treatments such as vaccinations and injections.

```python
# Plot the various HIV treatment methods
make_plot('nnrti_data.csv', 'NNRTI HIV Treatment Wholesale Markup Diagram', 'nnrti_visual')
make_plot('nrti_combin_data.csv', 'NRTI Combination HIV Treatment Wholesale Markup Diagram', 'nrti_combin_visual')
make_plot('pi_data.csv', 'Protease inhibitor HIV Treatment Wholesale Markup Diagram', 'pi_visual')
make_plot('other_data.csv', 'Other HIV Treatment Wholesale Markup Diagram', 'other_visual')
```

With that, we now have a generalizable method for visualizing information that is formatted with having Wholesale Acquisition Cost (WAC) data and the corresponding marked up Average Wholesale Price (AWP) data.

## Findings

Now that we have visualizations that can clearly communicate discrepancies within this step in the pharmaceutical cycle, just looking through the graphs the overall precedent appears to be all drugs are purchased at a lower price and marked up. The markup amount, however, seems to depend on the prescription that is being discussed. One extreme example is shown in the NNRTI graph which shows Neverapine which is purchased wholesale from vendors at an average of 10 dollars a prescription is sold at an average of 651 dollars per prescription. Per prescription is an important term in this markup, furthermore, as almost all of these drugs are sold at 30 doses a month distribution, which means that this markup price is not a one-off purchase but rather a monthly one that annually (in this example) would stack up to 7,812 dollars but was only purchased by the vendor initially at an annual wholesale price of 120 dollars which proportionally ends up being about a 6500% of the original price. What makes this even more insidious, furthermore, is that HIV treatment isn't an optional medication but rather is something that is quite literally keeping individuals alive. As a result of HIV-positive people's reliance on medicinal assistance, pharmaceutical companies can take advantage of a very vulnerable population that has no choice but to accept egregious price hikes that occur across the board.
In fact, all visualizations across all these datasets along with most found in the external research I did confirmed this result of increased prices that were large upcharges. Furthermore, it must be emphasized that this really is only a single step in this markup lifecycle. This analysis doesn't even begin to scratch other places where these charges are added or increased - particularly in the step between the drug production and the selling to vendors (which is arguably even larger of a hike up in certain cases). Consider, for example, a drug like Truvada - the most popular HIV prevention treatment that has a 99% efficacy in terms of preventing HIV transmission if taken consistently. As Daniel Summer indicates in the article linked below about Truvada and the cost markups, Truvada only costs about 6 dollars a month to manufacture but is sold at over 1500 dollars a prescription. In reality, this price increase is heavily alleviated by insurance coverages and couponing, but there still stands the issue that individuals without insurance and medical literacy (who are often the people that need the protection the most) cannot access this life-saving medicine due to its cost and the regulations associated (costly blood work, labs, check-ups, transportation to facilities, pharma deserts, etc).
